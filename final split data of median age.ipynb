```python
demo=pd.read_csv('D:\\pywork\\e-books\\Data-master\\python\\Demographics_State.csv')
```

```python
demo
```

```python
x=demo[['total_population','percent_white','percent_black','percent_asian','percent_hispanic','per_capita_income','median_rent']]
```

```python
y=demo[['median_age']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor=LinearRegression()
```

```python
regressor.fit(x_train,y_train)
```

```python
y_predictive_age =pd.DataFrame(regressor.predict(x_test),columns=['y_pred_median_age'])
```

```python
fig1=plt.figure()
ax1=fig1.add_subplot(111)
ax1.plot(y_actual)
ax1.plot(y_predictive_age)
```
